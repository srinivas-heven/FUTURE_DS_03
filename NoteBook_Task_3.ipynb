```python
# Importing the neccessary packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Loading CSV file into a Data Frame 
df = pd.read_csv("student_feedback.csv")
```

```python
# Checking Info about the Data Frame
df.info()
```

```python
# Looking at Top 5 Rows of the Data
df.head()
```

```python
# Droping the unwanted column as Data Frame have its own index
df.drop(columns="Unnamed: 0",inplace=True)
```

```python
df.iloc[:1,1:]
```

```python
# Creating a new column to find the overall rating of students
df["Overall Avg Rating"] = df.iloc[:,1:].mean(axis=1)
```

```python
# Rechecking DF after droping unwanted column
df.head()
```

```python
# Looking at the Statistics of the Data
df.describe()
```

```python
# Overall Raiting Average
df["Overall Avg Rating"].mean()
```

```python
# Creating a Data Frame which only have ratings
rating_cols = df.iloc[:,1:]
```

```python
# Finding out the average rating of the columns
avg_ratings = rating_cols.mean()
avg_ratings
```

```python
# Top 3 Feedback categories with hightest rating average
top_3_categories = (avg_ratings.sort_values(ascending=False).head(3)).round(2)
top_3_categories
```

```python
# Ploting a bar chart of Top 3 Feedback Categories
plt.figure(figsize=(10, 6))
top_3_categories.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 3 Feedback Categories with hightest rating average', fontsize=16)
plt.xlabel('Feedback Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

plt.show()
```

```python
# bottom 3 Feedback categories with lowest rating average
bottom_3_categories = (avg_ratings.sort_values(ascending=True).head(3)).round(2)
bottom_3_categories
```

```python
# Ploting a bar chart of bottom 3 Feedback Categories
plt.figure(figsize=(10, 6))
bottom_3_categories.plot(kind='barh', color='skyblue', edgecolor='black')

plt.title('Bottom 3 Feedback Categories with lowest rating average', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Feedback Category', fontsize=12)
plt.tight_layout()

plt.show()
```

```python
# Visualize the distribution and mean of each rating column in the DataFrame using custom horizontal bar plots.

for i in df.columns[1:]:
    fig = plt.figure(figsize=(20,4))
    gridspec.GridSpec(1,4)

    plt.subplot2grid((1,4),(0,0)).set_facecolor("w")
    plt.annotate(round(df[i].mean(),1),(0.2,0.2),fontsize=100,color="k")
    plt.yticks([])
    plt.xticks([])

    col1=df[i].value_counts().sort_index()
    plt.subplot2grid((1,4),(0,1),colspan=3,rowspan=1).set_facecolor("w")
    plt.barh(col1.index,(max(col1))*0.3+max(col1),color="lightgrey",height=0.5)
    plt.barh(col1.index,col1,height=0.5,color="mediumseagreen",ec="g")
    plt.xticks([])
    plt.yticks(fontsize=15,color="k")
    plt.xlabel("# of students")
    plt.ylabel("Score")
    plt.title("<----- "+i+" ----->",color="k",fontsize=20)
    for x,y in enumerate(col1,min(col1.index)):
        plt.annotate("( "+str(y)+" )",(y+y*0.01,x-0.1))
    plt.show()
    print("-"*120)
    
```

- **Scale**: Most scores are between 1–10.
